```python
import cv2
import glob
from tqdm import tqdm
from albumentations import HorizontalFlip, VerticalFlip, RandomRotate90
import matplotlib.pyplot as plt
import os
import math
```

```python

jpgs = glob.glob('/content/gdrive/MyDrive/Planes/600Images/jpgs/*.jpg')
pngs = glob.glob('/content/gdrive/MyDrive/Planes/600Images/pngs/*.png')

for x,y in zip(sorted(jpgs), sorted(pngs)):
  if x[:-5] != 'x':

    base = x[-10:-7]
    subase =x[-6:-4]
    print(base)
    print(subase)
    print('-----------------------')
    x=cv2.imread(x, cv2.IMREAD_COLOR)

    y=cv2.imread(y, cv2.IMREAD_COLOR)

    aug = HorizontalFlip(always_apply=False, p=1.0)
    augmented = aug(image=x, mask=y)
    x1=augmented['image']
    y1=augmented['mask']

    aug = VerticalFlip(always_apply=False, p=1.0)
    augmented = aug(image=x, mask=y)
    x2=augmented['image']
    y2=augmented['mask']

    aug = RandomRotate90(always_apply=False, p=1.0)
    augmented = aug(image=x, mask=y)
    x3=augmented['image']
    y3=augmented['mask']

    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/jpgs/{x}_{y}_0x.jpg'.format(x = base, y = subase), x1)
    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/pngs/{x}_{y}_0x.png'.format(x = base, y = subase), y1)

    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/jpgs/{x}_{y}_1x.jpg'.format(x = base, y = subase), x2)
    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/pngs/{x}_{y}_1x.png'.format(x = base, y = subase), y2)

    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/jpgs/{x}_{y}_2x.jpg'.format(x = base, y = subase), x3)
    cv2.imwrite('/content/gdrive/MyDrive/Planes/600Images/pngs/{x}_{y}_2x.png'.format(x = base, y = subase), y3)



```
